```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_large = pd.read_csv("Largest.csv")
df_small = pd.read_csv("Smallest_partial.csv")
```

```python

plt.figure(figsize=(15,8))
plt.title("Loss Comparison", size=20)
sns.lineplot(data=df_large, x='Step', y='Value', color='r', alpha=0.5, label="ViTB/32@224")
sns.lineplot(data=df_small, x='Step', y='Value', color='darkblue', alpha=0.5, label="ViTB/16@112")

plt.legend(fontsize=15)
plt.xlim(0, max(df_large['Step']))
plt.xlabel("Step", fontsize=15)
plt.ylabel("Loss", fontsize=15)
plt.savefig("loss_analysis.png")
```
